## Data Cleaning Process

### Join all datasets, and create new Dataframe.

```python
# merge data, and create new df

df = [Licen_exam2017_df,Licen_exam2018_df,Licen_exam2019_df,
      Licen_exam2020_df,Licen_exam2021_df,Licen_exam2022_df]
df_new = pd.concat(df)
```

```python
df_new.head(20)
```

```python
#create date col
df_new['Date'] = pd.to_datetime(df_new[['Year','Month']].assign(Day=1))

df_new.head()
```

```python
#create new df that use date instead of year, month.
df = df_new[['Date', 'Licensee Brokers', 'Licensee Salespersons',
       'Total Licensee', 'REL Broker', 'REL Salesperson',
       'Examinations brokers', 'Examinations Sales']]
df.head()
```

### Group data by Fiscal year.

```python
# For federal gov, FY runs from 0ct. 1 to Sept. 30.
df['Fiscal_Year'] = df['Date'].dt.to_period('Q-SEP').dt.qyear

df.head(25)
```

```python
##Drop FY 2017 data, and only group FY 2018-2022.
df_FY = df[df['Fiscal_Year'] != 2017]

df_FY = df_FY.drop('Date', axis=1)
df_FY = df_FY[['Fiscal_Year','Licensee Brokers', 'Licensee Salespersons',
       'Total Licensee', 'REL Broker', 'REL Salesperson',
       'Examinations brokers', 'Examinations Sales']]
```

## Data Insights

```python
df_FY.head()
```

```python
# Calculate total examees of brokers based on each FY.
Ex_brokers_by_fy = df_FY.groupby('Fiscal_Year').sum()['Examinations brokers'].map('{:,.0f}'.format)
print(Ex_brokers_by_fy)
print('------------------')

# Avg examees of brokers based on each FY
Ag_exbrokers_by_fy = df_FY.groupby('Fiscal_Year').mean()['Examinations brokers'].map('{:.2f}'.format)
print(Ag_exbrokers_by_fy)
```

```python
# Calculate total examinees of sales based on each FY.
Ex_sales_by_fy = df_FY.groupby('Fiscal_Year').sum()['Examinations Sales'].map('{:,.0f}'.format)
print(Ex_sales_by_fy)
print('---------------')

# Avg examees of sales based on each FY
Ag_exsales_by_fy = df_FY.groupby('Fiscal_Year').mean()['Examinations Sales'].map('{:,.2f}'.format)
print(Ag_exsales_by_fy)
```
